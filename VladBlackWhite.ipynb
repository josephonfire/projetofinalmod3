```python
def para_preto(imagem_cinza, limiar=127):
    altura, largura = imagem_cinza.shape
    imagem_preto = np.zeros((altura, largura), dtype=np)
    for i in range(altura):
        for j in range(largura):
            if imagem_cinza[i, j] > limiar:
                imagem_preto[i, j] = 255
            else:
                imagem_preto[i, j] = 0
    return imagem_preto
```

```python
imagem_preto = para_preto(imagem_cinza)
cv2_imshow(imagem_preto)
```
